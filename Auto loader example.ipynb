```python
df = spark.readStream\
      .format("cloudFiles")\
      .option("cloudFiles.format", "json")\
      .option("cloudFiles.schemaLocation", "/Volumes/dev_ecommerce/bronze/ecommerce_data/orders/schema")\
      .option("cloudFiles.inferColumnTypes", "true")\
      .option("cloudFiles.schemaEvolutionMode", "addNewColumns")\
      .option("cloudFiles.rescuedDataColumn", "_rescued_data")\
      .load("/Volumes/dev_ecommerce/bronze/ecommerce_data/orders/raw/")
```

```python
df.writeStream\
    .format("delta")\
    .option("checkpointLocation", "/Volumes/dev_ecommerce/bronze/ecommerce_data/orders/_checkpoints/")\
    .trigger(availableNow=True)\
    .toTable("dev_ecommerce.bronze.orders_bronze")
```
